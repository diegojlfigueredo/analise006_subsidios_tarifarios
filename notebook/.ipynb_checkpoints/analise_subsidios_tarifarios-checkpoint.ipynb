```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Caminhos
caminho_csv = r'../dados/subsidios-tarifarios.csv'
caminho_imagens = r'../imagens'

# 1. Leitura dos dados
df = pd.read_csv(caminho_csv, encoding='latin1', sep=';')

# 2. Ajustes de colunas
df.columns = df.columns.str.strip()

# 3. Convertendo datas
df['DatSubsidio'] = pd.to_datetime(df['DatSubsidio'], errors='coerce', dayfirst=True)

# 4. Estatísticas descritivas
print("===== Estatísticas descritivas =====")
print(df.describe(include='all'))

# 5. Verificando valores nulos
print("\n===== Valores nulos =====")
print(df.isnull().sum())

# 6. Visualização da distribuição do subsídio
plt.figure(figsize=(10, 6))
sns.histplot(df['VlrSubsidio'], bins=30, kde=True, color='skyblue')
plt.xlabel('Valor do Subsídio')
plt.ylabel('Frequência')
plt.title('Distribuição dos Valores de Subsídios')
plt.tight_layout()
plt.savefig(os.path.join(caminho_imagens, 'distribuicao_valor_subsidio.png'))
plt.close()

# 7. Subsídios por tipo
plt.figure(figsize=(12, 6))
tipo_counts = df['DscTipoSubsidio'].value_counts()
sns.barplot(x=tipo_counts.index, y=tipo_counts.values, palette='Set2')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Tipo de Subsídio')
plt.ylabel('Quantidade')
plt.title('Quantidade de Subsídios por Tipo')
plt.tight_layout()
plt.savefig(os.path.join(caminho_imagens, 'subsidiostipo.png'))
plt.close()

# 8. Subsídios por agente
plt.figure(figsize=(12, 6))
agente_counts = df['SigAgente'].value_counts().head(10)
sns.barplot(x=agente_counts.index, y=agente_counts.values, palette='Set3')
plt.xlabel('Agente')
plt.ylabel('Quantidade')
plt.title('Top 10 Agentes com mais Subsídios')
plt.tight_layout()
plt.savefig(os.path.join(caminho_imagens, 'subsidioporagente.png'))
plt.close()

# 9. Evolução temporal dos subsídios
df['Ano'] = df['DatSubsidio'].dt.year
evolucao = df.groupby('Ano')['VlrSubsidio'].sum()
plt.figure(figsize=(10, 6))
evolucao.plot(kind='line', marker='o', color='green')
plt.xlabel('Ano')
plt.ylabel('Total de Subsídios')
plt.title('Evolução Temporal dos Subsídios')
plt.tight_layout()
plt.savefig(os.path.join(caminho_imagens, 'evolucao_temporal.png'))
plt.close()

print("✅ Análise concluída! Gráficos salvos na pasta de imagens.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Caminho correto para o arquivo CSV
caminho_csv = r'D:\03 - Portfolios\portfolio_diego\analises_de_dados\subsidios-tarifarios\dados\subsidios-tarifarios.csv'

# Lendo o CSV com todas as colunas como texto para evitar problemas de tipo
df = pd.read_csv(caminho_csv, encoding='latin1', sep=';', dtype=str, low_memory=False)

# Criação da pasta de saída para as imagens
pasta_imagens = r'D:\03 - Portfolios\portfolio_diego\analises_de_dados\subsidios-tarifarios\imagens'
os.makedirs(pasta_imagens, exist_ok=True)

print(df.head())
```

```python
# Exibir as colunas e uma amostra inicial
print("Colunas disponíveis:")
print(df.columns)

# Exibir as primeiras linhas do dataframe
print("\nAmostra dos dados:")
print(df.head())
```

```python
# Convertendo a coluna de valor de subsídio para numérico
df['VlrSubsidio'] = pd.to_numeric(df['VlrSubsidio'].str.replace(',', '.'), errors='coerce')

# Convertendo datas
datas = ['DatGeracaoConjuntoDados', 'DatSubsidio', 'DatAssinaturaAto', 
         'DatInicioVigenciaAto', 'DatFimVigenciaAto', 'DthPublicacaoAto']

for col in datas:
    df[col] = pd.to_datetime(df[col], errors='coerce', dayfirst=True)

print("\nTipos de dados após conversão:")
print(df.dtypes)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Define a pasta onde vamos salvar as imagens
pasta_imagens = r"D:\03 - Portfolios\portfolio_diego\analises_de_dados\subsidios-tarifarios\imagens"
os.makedirs(pasta_imagens, exist_ok=True)

# Cria o histograma
plt.figure(figsize=(10, 6))
sns.histplot(df['VlrSubsidio'], bins=50, kde=True, color='skyblue')
plt.xlabel('Valor do Subsídio')
plt.ylabel('Frequência')
plt.title('Distribuição dos Valores de Subsídio')

# Salva a imagem
caminho_imagem = os.path.join(pasta_imagens, 'histograma_valores_subsidio.png')
plt.savefig(caminho_imagem, dpi=300, bbox_inches='tight')
plt.close()

print(f"Histograma salvo em: {caminho_imagem}")
```

```python
# Agrupamento por agente
top10_agentes = df.groupby(['NomAgente', 'SigAgente'])['VlrSubsidio'].sum().reset_index()

# Ordena em ordem decrescente
top10_agentes = top10_agentes.sort_values(by='VlrSubsidio', ascending=False).head(10)

# Cria o gráfico
plt.figure(figsize=(12, 8))
sns.barplot(
    x='VlrSubsidio',
    y='NomAgente',
    data=top10_agentes,
    palette='Blues_d'
)
plt.xlabel('Total de Subsídios')
plt.ylabel('Agente')
plt.title('Top 10 Agentes que Mais Receberam Subsídios')

# Salva a imagem
caminho_imagem = os.path.join(pasta_imagens, 'top10_agentes_subsidios.png')
plt.savefig(caminho_imagem, dpi=300, bbox_inches='tight')
plt.close()

print(f"Gráfico Top 10 agentes salvo em: {caminho_imagem}")
```

```python
# Agrupando por Ano e Mês
df['AnoMes'] = df['DatSubsidio'].dt.to_period('M')
evolucao_subsidios = df.groupby('AnoMes')['VlrSubsidio'].sum().reset_index()
evolucao_subsidios['AnoMes'] = evolucao_subsidios['AnoMes'].astype(str)

# Plotando a série temporal
plt.figure(figsize=(14, 6))
sns.lineplot(
    data=evolucao_subsidios,
    x='AnoMes',
    y='VlrSubsidio',
    marker='o',
    color='steelblue'
)
plt.xticks(rotation=45)
plt.xlabel('Ano-Mês')
plt.ylabel('Total de Subsídios')
plt.title('Evolução Temporal dos Subsídios')

# Salva a imagem
caminho_imagem = os.path.join(pasta_imagens, 'evolucao_subsidios.png')
plt.savefig(caminho_imagem, dpi=300, bbox_inches='tight')
plt.close()

print(f"Gráfico de evolução temporal salvo em: {caminho_imagem}")
```

```python
# Agrupando por tipo de subsídio
tipos_subsidio = df.groupby('DscTipoSubsidio')['VlrSubsidio'].sum().sort_values(ascending=False).reset_index()

# Exibindo os top 10 tipos de subsídio
top_tipos_subsidio = tipos_subsidio.head(10)

# Plotando o gráfico
plt.figure(figsize=(12, 6))
sns.barplot(
    data=top_tipos_subsidio,
    x='VlrSubsidio',
    y='DscTipoSubsidio',
    palette='magma'
)
plt.xlabel('Valor Total de Subsídios')
plt.ylabel('Tipo de Subsídio')
plt.title('Top 10 Tipos de Subsídio')

# Salvar imagem
caminho_imagem = os.path.join(pasta_imagens, 'top10_tipos_subsidio.png')
plt.savefig(caminho_imagem, dpi=300, bbox_inches='tight')
plt.close()

print(f"Gráfico dos tipos de subsídio salvo em: {caminho_imagem}")
```

```python
# Agrupando por ato normativo
atos_normativos = df.groupby('DscAtoNormativo')['VlrSubsidio'].sum().sort_values(ascending=False).reset_index()

# Exibindo os top 10 atos normativos
top_atos_normativos = atos_normativos.head(10)

# Plotando o gráfico
plt.figure(figsize=(12, 6))
sns.barplot(
    data=top_atos_normativos,
    x='VlrSubsidio',
    y='DscAtoNormativo',
    palette='coolwarm'
)
plt.xlabel('Valor Total de Subsídios')
plt.ylabel('Ato Normativo')
plt.title('Top 10 Atos Normativos por Valor de Subsídios')

# Salvar imagem
caminho_imagem = os.path.join(pasta_imagens, 'top10_atos_normativos.png')
plt.savefig(caminho_imagem, dpi=300, bbox_inches='tight')
plt.close()

print(f"Gráfico dos atos normativos salvo em: {caminho_imagem}")
```

```python
# Agrupar por data (DatSubsidio) e somar os valores
df_temporal = (
    df.groupby(df['DatSubsidio'].dt.to_period('M'))['VlrSubsidio']
    .sum()
    .reset_index()
)
df_temporal['DatSubsidio'] = df_temporal['DatSubsidio'].dt.to_timestamp()

# Plotar o gráfico de linha
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=df_temporal,
    x='DatSubsidio',
    y='VlrSubsidio',
    marker='o'
)
plt.xlabel('Data')
plt.ylabel('Valor Total de Subsídios')
plt.title('Evolução Temporal dos Subsídios')

# Salvar imagem
caminho_imagem = os.path.join(pasta_imagens, 'evolucao_subsidios.png')
plt.savefig(caminho_imagem, dpi=300, bbox_inches='tight')
plt.close()

print(f"Gráfico de evolução temporal salvo em: {caminho_imagem}")
```
